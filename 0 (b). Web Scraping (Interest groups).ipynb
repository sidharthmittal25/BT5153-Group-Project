```python
import pandas as pd
import warnings
warnings.filterwarnings('ignore')

df = pd.read_csv('2019lender.csv')
del df['Unnamed: 0']
df = df.replace('0',0)
df
```

```python
import requests
from bs4 import BeautifulSoup as bs

link_list = list(df.link)

for i, j in zip(link_list[80789:83030], range(80789,83030)):  #number to be replaced by the start & end index assigned

    URL = i

    # opening up connection, grabbing the page
    html_text = requests.get(URL).text

    # html parsing
    soup = bs(html_text, 'lxml')

    # grabs team
    jobs = soup.find_all('article', class_ = 'teamCard default compact vertical')
    
    if jobs == []:
        team = 'no_team'
    else: # grabs name of each team
        team = []
        for job in jobs:
            name = job.find('h1', class_='name').a['title']
            team.append(name)
    
    df['team'][j] = team
    print(j, team) # to visualize progress
    
    # export to csv for every 1000 record scrapped
    if j in range(0, 83030, 1000): #number to be replaced by the start & end index assigned
        df.to_csv("2019lender_"+str(j)+".csv")
    
```
